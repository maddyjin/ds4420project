### Merge senator_ideologies_df onto all_senators_with_bios

```python
# imports
import pandas as pd
```

```python
# load the two dataframes
senators_df   = pd.read_csv('../../data/senators/all_senators_with_bios.csv')
ideologies_df = pd.read_csv('../../data/ideology_scores/senator_ideologies.csv')
```

```python
# left join — every senator row is kept; ideology columns are null for senators
# who predate both scoring systems (very early congresses) or are missing from both sources
senators_df = senators_df.merge(
    ideologies_df[['bioguide_id', 'nominate_dim1', 'nominate_dim2', 'govtrack_ideology_score']],
    left_on='id',
    right_on='bioguide_id',
    how='left'
).drop(columns=['bioguide_id'])  # redundant with 'id'
```

```python
# how many senators got ideology scores?
scored = senators_df['nominate_dim1_x'].notna().sum()
total  = len(senators_df)
print(f'{scored} of {total} senators have a NOMINATE score ({scored/total:.1%})')

senators_df.head()
```

```python
# overwrite the senators file with ideology columns included
senators_df.to_csv('../../data/senators/all_senators_with_bios.csv', index=False)
```
