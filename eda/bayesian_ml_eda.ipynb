```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in cleaned csv
df = pd.read_csv("../data/all_bills_cleaned.csv")
df.head()
df.shape
```

```python
# 2b. Cosponsor distribution
sns.histplot(df["Number of Cosponsors"], bins=30)
plt.title("Distribution of Cosponsors")
plt.show()
```

```python
# 2c. Top committees
df["Committees"].value_counts().head(10).plot(kind="barh")
plt.title("Top Committees")
plt.show()
```

```python
# 2d. Extract party from sponsor
df["party"] = df["Sponsor"].str.extract(r"\[(?:Rep|Sen)\.-([DRI])")
df["party"].value_counts().plot(kind="bar")
plt.title("Sponsor Party Distribution")
plt.show()
```

```python
# 2e. Create binary outcome: passed or not
df["passed"] = df["Latest Action"].str.contains(
    "Became Public Law|Signed by President|Passed", na=False
)
df["passed"].value_counts().plot(kind="bar")
plt.title("Bill Passage Distribution")
plt.show()
```

```python
# 2f. Cosponsors vs passage
sns.boxplot(x="passed", y="Number of Cosponsors", data=df)
plt.title("Cosponsors vs Bill Passage")
plt.show()
```

```python
# 2g. Party vs passage
sns.countplot(x="party", hue="passed", data=df)
plt.title("Party vs Passage")
plt.show()
```
